# **Wikipedia**

**Wikipedia is a free content, multilingual online encyclopedia written and maintained by a community of volunteers through a model of open collaboration, using a wiki-based editing system. As of February 2022, the site is available in 325 different languages, has over 300,000 active editors and contains over 6 million English articles.**

<sup>Source: [Article on Wikipedia](https://en.wikipedia.org/wiki/Wikipedia) from Wikipedia.org<sup/>

## **Summary of a Wiki Article**

```python
wikipedia.summary('grace hopper', auto_suggest=False)
```

## **Section within a Wiki Article**

```python
wikipedia.page('grace hopper', auto_suggest=False).section('COBOL')
```

# **References within a Wiki Article**

```python
wikipedia.page('grace hopper', auto_suggest=False).references
```

## **Retrieving images from an article**

```python
wikipedia.page('grace hopper', auto_suggest=False).images
```

```python
display.Image(wikipedia.page('grace hopper', auto_suggest=False).images[10])
```

## **Retrieve URL of an article**

```python
wikipedia.page('grace hopper', auto_suggest=False).url
```

## **Categories of the article**

```python
wikipedia.page('grace hopper', auto_suggest=False).categories
```

# **Changing the default language setting**

```python
wikipedia.languages()
```

```python
wikipedia.set_lang('es')

resumen_de_informatica = wikipedia.summary('Ciencias de la computación', auto_suggest=False)

print(resumen_de_informatica)
```

```python
wikipedia.set_lang('en')
wikipedia.summary('computer science', auto_suggest=False)
```
