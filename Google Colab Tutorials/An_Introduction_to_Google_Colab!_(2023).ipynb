# **An Introduction to Google Colab!**

Colaboratory, or “Colab” for short, is a product from Google Research. Colab allows users to write and execute arbitrary python code through the browser. Colab is a hosted Jupyter notebook service that requires no setup to use, while providing access free of charge to computing resources including GPUs.

***Important***

In order to use Google Colab you need to be signed into a Google account

## **How Long Do Colab Runtimes Last?**

A runtime is the length of your Google Colab session. Once a runtime shuts down all the data from the Colab page is erased (assigned Python variables, data output from code, models, data downloaded into the content folder).

Runtimes will time out if a user is idle. In the free-of-charge version of Colab, notebooks can run for at most 12 hours.

<sup>Source: [Colaboratory Frequently Asked Questions](https://research.google.com/colaboratory/intl/en-GB/faq.html)</sup>

## **GPU Runtimes**

A graphics processing unit (GPU) is specialized hardware that performs certain computations much faster than a traditional computer's central processing unit (CPU). Python machine learning modules such as TensorFlow and Torch utilize GPUs.

The types of GPUs that are available in Colab vary over time. This is necessary for Colab to be able to provide access to these resources free of charge.

***Important***

If you are working with a non-GPU runtime and you switch to a GPU runtime you will lose all your data.

<sup>Source: [GPUs for Machine Learning](https://itconnect.uw.edu/guides-by-topic/research/research-computing/gpus-for-machine-learning/) from the University of Washington's Research Computing Page</sup>

```python
from diffusers import DiffusionPipeline
import torch

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")

# if using torch < 2.0
# pipe.enable_xformers_memory_efficient_attention()

prompt = "An astronaut riding a green horse"

images = pipe(prompt=prompt).images[0]
```

## **Writing in Markdown**

- Using hash tags sets sections in the notebook:
 - \# or \<h1>\</h1> for Heading Level 1
 - \## or \<h2>\</h2> for Heading Level 2
 - \### or \<h3>\</h1> for Heading Level 3
 - \#### or \<h4>\</h4> for Heading Level 4
 - \##### or \<h5>\</h5> for Heading Level 5
 - \###### or \<h6>\</h6> for Heading Level 6


- Text is **bolded** with the text in between two asterisks (\**   \**) or with a double underscore (\__ \__)

- Text is *italicized* with the text in between singe asterisks (\* \*) or between underscores (\_ \_)

- Text is ***bolded and italicized*** with the text in between three asterisks (\***   \***)

- Text can be styled in code blocks: `import random`
 - This is done by putting text between backticks (\` \`).



## **Equations with LaTeX**

- LaTeX can be rendered by putting text between dollar signs (\$ \\$):
 - Example 1: $\Large{\frac{x}{2}=10}$
 - Example 2: $\int_{a}^{b} 3x^2 \,dx$
 - Example 3: $\vec{p} \cdot \vec{q}$
