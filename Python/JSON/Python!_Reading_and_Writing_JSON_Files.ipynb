# **JSON**

**JSON (JavaScript Object Notation) is a lightweight data-interchange format based on a subset of the JavaScript Programming Language. The text format is independent of language, meaning JSON files can be read or written using Python, R, C++, Rust and many other programming languages.** 

**JSON is widely used for interacting with application programming interfaces (APIs) on the web.**

**There are two structures that JSON is built upon: an ordered list with values, and a collection of name and value pairs. Within Python these two structures correspond to lists and dictionaries.**

## **Reading a JSON file**

```python
with open('playlist_data.json', 'r') as read_file:
    #the json file is read into a nested dictionary
    playlist_dict = json.load(read_file)
```

```python
playlist_dict.keys()
```

```python
playlist_dict['info']
```

```python
type(playlist_dict['playlists'])
```

### **Parsing the JSON data**

```python
#iterate through the playlists from the json file
for paylist in playlist_dict['playlists']:
    print(paylist['name'])
```

```python
for idx,paylist in enumerate(playlist_dict['playlists']):
    if paylist['name'] == 'Rap Classics':
        print(idx)
```

```python
rap_dict = playlist_dict['playlists'][170]
rap_dict
```

```python
for song_data in rap_dict['tracks']:
    print(f'Artist: {song_data["artist_name"]} Song: {song_data["track_name"]}')
```

## **Writing a JSON file**

```python
with open('rap_classics.json', 'w') as outfile:
    json.dump(rap_dict, outfile)
```
