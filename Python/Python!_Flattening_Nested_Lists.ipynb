# **Lists**

**Within Python, lists are a mutable, dynamic class that can reference multiple elements. Python lists are pointer structures, meaning they reference the data rather than store it. The advantages of pointer structures are they do not require sequential storage space and the size of pointer structures can be dynamic.**

**Lists can reference different data types within their structure.**

**Lists are similar to arrays in other languages such as Java or C++. However, Python lists are more flexible than arrays, because the size of lists can be altered after creation and lists have more operations that can be executed relative to arrays.**

**There are drawbacks to Python lists. Lists take up more memory space than arrays and conducting operations on lists takes longer than it would with arrays.**

# **Nested List**

**Nested lists are Python lists within a list. Nested lists are used to store data such as pixel values for an image or stock prices for companies across multiple dates.**

```python
#initialize an empty list
nest_list = []

#create a nested list by appending 10,000 randomly generated lists to our nest_list
while len(nest_list) < 10000:
    nest_list.append(np.random.randint(0,1000,size=100).tolist())
```

```python
for sublist in nest_list:
  print(sublist)
```

```python
#print out the type of element within our nest_list to confirm they are all sub lists
for sub_list in nest_list:
    print(type(sub_list))
```

# **Flattening Nested Lists**

## **Method 1: `for` Loops**

```python
def for_flat(nested_list):

  '''Function that flattens a nested list using for loops.

  Parameters
  ----------
  nested_list: The input is a list where the underlying elements are lists.
  This is termed a nested list.

  Returns
  -------
  flat_list: Returns a flattened list.
  '''

  #create an empty list that will store the underlying elements from the sub lists
  flat_list = []
  for sublist in nested_list:
      for element in sublist:
          flat_list.append(element)

  return flat_list
```

```python
flat_list_1 = for_flat(nest_list)
flat_list_1
```

## **Method 2: List Comprehension**

```python
def list_comp_flat(nested_list):

  '''Function that flattens a nested list using list comprehension.

  Parameters
  ----------
  nested_list: The input is a list where the underlying elements are lists.
  This is termed a nested list.

  Returns
  -------
  flat_list: Returns a flattened list.
  '''

  return [element for sublist in nested_list for element in sublist]
```

```python
flat_list_2 = list_comp_flat(nest_list)
```

## **Method 3: itertools**

```python
def iter_flat(nested_list):

  '''Function that flattens a nested list using the chain function that returns
     elements from the first iterable until it is exhausted, then proceeds to
     the next iterable, until all of the iterables are exhausted.

  Parameters
  ----------
  nested_list: The input is a list where the underlying elements are lists.
  This is termed a nested list.

  Returns
  -------
  flat_list: Returns a flattened list.'''

  return list(itertools.chain(*nested_list))
```

```python
flat_list_3 = iter_flat(nest_list)
```

<sup><sup/>
Source: [itertools Documentation](https://docs.python.org/3/library/itertools.html#itertools.chain)

## **Method 4: `extend` function**

```python
def extend_flat(nested_list):

  '''Function that flattens a nested list using the extend function.

  Parameters
  ----------
  nested_list: The input is a list where the underlying elements are lists.
  This is termed a nested list.

  Returns
  -------
  flat_list: Returns a flattened list.'''

  #create an empty list that will store the underlying elements from the sub lists
  flat_list = []

  #extend the list by appending elements from an iterable
  for sublist in nested_list:
      flat_list.extend(sublist)

  return flat_list
```

```python
flat_list_4 = extend_flat(nest_list)
```

<sup><sup/>
Source: [More on Lists](https://docs.python.org/3/tutorial/datastructures.html#more-on-lists) from python.org

## **Method 5: functools**

```python
def func_flat(nested_list):

  '''Function that flattens a nested list using the functools module.
  The reduce function is used to reduce an iterable to a single value.
  The operator.iconcat combines all the underlying elements within the
  sublists together which are then saved to a single list.



  Parameters
  ----------
  nested_list: The input is a list where the underlying elements are lists.
  This is termed a nested list.

  Returns
  -------
  flat_list: Returns a flattened list.'''

  return functools.reduce(operator.iconcat, nested_list, [])
```

```python
flat_list_5 = func_flat(nest_list)
```

### **Check that all lists have been flattened**

```python
flat_list_1 == flat_list_2 == flat_list_3 == flat_list_4 == flat_list_5
```

## **Comparing the Time Complexity of Flattening Methods**

```python
#takes approximately 3-5 minutes to run

timer_dict = {'For Loop':[],'List Comprehension':[],
              'itertools':[],'extend Function':[],
              'functools':[]}

for iteration in range(0,260000,5000):
    temp_list = []
    while len(temp_list) < iteration:
        temp_list.append(np.random.randint(0,1000,size=100).tolist())

    start = time.time()
    for_flat(temp_list)
    end = time.time()
    timer_dict['For Loop'].append(end-start)

    start = time.time()
    list_comp_flat(temp_list)
    end = time.time()
    timer_dict['List Comprehension'].append(end-start)

    start = time.time()
    iter_flat(temp_list)
    end = time.time()
    timer_dict['itertools'].append(end-start)

    start = time.time()
    extend_flat(temp_list)
    end = time.time()
    timer_dict['extend Function'].append(end-start)

    start = time.time()
    func_flat(temp_list)
    end = time.time()
    timer_dict['functools'].append(end-start)
```

```python
for key,timer in timer_dict.items():
    plt.plot(range(0,260000,5000), timer, label=key)

plt.ylabel('Time to Run in Seconds')
plt.xlabel('Number of Elements in Nested List')
plt.gca().xaxis.set_major_formatter('{x:,.0f}')
plt.legend();
```
