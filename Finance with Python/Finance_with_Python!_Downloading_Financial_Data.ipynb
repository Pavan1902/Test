# **yfinance**

**yfinance is a Python module created after Yahoo! Finance's API was discontinued. The module allows users to download historical market data from Yahoo! Finance.**

<sup>Source: [Reliably download historical market data from Yahoo! Finance with Python](https://aroussi.com/post/python-yahoo-finance) by Ran Aroussi<sup/>

## **Downloading Historial Stock and Index Data**

```
#periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
#intervals: 1m,2m,5m,15m,30m,60m,90m,1h,1d,5d,1wk,1mo,3mo

nasdaq_100 = yf.download(tickers='qqq',period='1y',interval='1mo',rounding=True)
```

```
nasdaq_100
```

## **Downloading Historical Prices for Multiple Stocks**

```
etf_data = yf.download(tickers='qqq spy dia',period='1y',interval='1mo',rounding=True)
```

```
etf_data
```

```
etf_data['Adj Close']
```

## **Downloading Company Financial Data**

```
aapl = yf.Ticker('AAPL')
```

```
aapl.get_financials()
```

## **Retrieve Financial News**

```
aapl.get_news()
```

## **Retrieve Financial Analys Recommendations**

```
aapl.recommendations
```
