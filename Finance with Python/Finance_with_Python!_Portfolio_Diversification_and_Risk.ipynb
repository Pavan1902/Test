# **What is Diversification?**

**Within finance and portfolio management, diversification is when different assets are held in the portfolio so that the exposure to
any particular asset is limited. Diversification can limit exposure to a particular industry, country or asset class. For example if a stockholder only owns shares of technology companies (Apple, Microsoft, Nvidia, Netflix), they can limit their exposure to the tech sector by purchasing bank shares (JP Morgan and Bank of America) and in utility companies (Dominion Energy and Duke).**

### **Daily Holding Period Returns**

**To compare the returns of assets, portfolio managers often calculate the holding period returns rather than the price difference of an asset**

**In our case, we will calculate the daily holding period returns of the price of an asset's previous day's close by the current price of the asset.**

$\LARGE{\text{Daily Holding Period Return} = \frac{\text{Current Price of Asset}}{\text{Previous Closing Price of Asset}}}-1$

<br/>

$\Large{\text{Example: AAPL Current Price = \$178, AAPL Previous Closing Price = \$177}}$

$\Large{\text{AAPL Daily Holding Period Return = }\frac{\text{\$178}}{\text{\$177}}-1 =0.565\%}$

### **Historical Volatility (Standard Deviation) of Holding Period Returns**

**Portfolio managers calculate the volatility of a portfolio by taking the standard deviation of returns. Standard deviation is the dispersion of returns; the higher the standard deviation for an asset's returns the more volatile and risky it is (large and quick price swings). Lower the standard deviation leads to less volatility (small and measured price swings).**

**For example, we can look at the volatility of a single stock, in our case Apple. If we calculate the recent volatility for Apple returns and determine the daily standard deviation is 1.8%, then we can estimate on any given day Apple shares will most likely returns somewhere betwee +1.8% and -1.8%.** 

**There are other ways traders calculate volatility, most notably market makers use implied volatility from options to forecast the potential price movement of assets.**

<sup>[Discover Options Volatility](https://www.cmegroup.com/education/courses/introduction-to-options/discover-options-volatility.html) from the CME Group<sup/>

$\LARGE{s = \sqrt{\frac{(x_i-\bar{x})}{n-1}}}$

<br/>

$\text{Where: $s$ is the sample standard deviation, $x_i$ are each of the individual returns for the asset, $\bar{x}$ is the average return for the asset and $n$ is the total number of returns in the data set.}$

```python
stdev_history = []

for i in range(1,len(stock_histories_filtered)+25,25):
  stdev_history.append([i,sum([stock['Close'] for stock in stock_histories_filtered[:i]]).pct_change(1).std()])
```

```python
plt.plot([num[0] for num in stdev_history], [stdev[1] for stdev in stdev_history],
         marker = 'o', markersize=9, markeredgecolor = 'red')

plt.xlabel('Number of Stocks in Portfolio')
plt.ylabel('Daily Standard Deviation (Portfolio Volatility)')
plt.title('Price Weighted Equity (Stock) Portfolio')
plt.xticks(ticks=range(0,550,50));
```

## **Volatility by Asset Class**

- **Historically, equities have been a risky asset class potentially giving higher rates of return, but with significant downside risk during recessions or periods of uncertainty.**


- **Bonds have been less risky because they are paid first in the capital structure for a company, meaning the loss of principal is far less likely relative to equities. With less risk bonds usually have lower returns over time relative to equities.**

- **Commodities tend to be a volatile asset class because of the underlying commodities themselves, but also because of the instruments used to access the market. The instruments that back most commodities funds are derivatives, usually futures and options. Derivatives are volatile instruments, combined with the volatility of commodities such as oil the price swings for these funds can be substantial.**

- **Real estate investments have similar risk/return profiles to equities. Often retail investors can access the real estate market through real estate investment trusts (REITs) which are liquid and trade on financial exchanges. Larger investor can directly purchase buildings or infrastructure to invest in real estate. REITs share similar risks to equities; if there is a recession then the apartment buildings owned by REITs may not generate as much cashflow putting their dividend distribution at risk. Privately owned buildings share the same risk and have the additional illiquidity risk.**

- **Given that cryptocurrencies are a new asset class, they carry the most risk out of any other asset class listed here. Not only is the asset itself novel, but regulations are still being written for cryptocurrencies. This layers on additional risk of fraud. However, some of the cryptocurrencies have netted a hefty profit for investors willing to take the risk.**

### **Annualized Daily Volatility**

```python
asset_vol = [252**.5 * asset.pct_change(1).std() for asset in asset_list]

for title,vol in zip(asset_class_df.columns,asset_vol):
  plt.bar(x=title,height=vol)

for i, v in enumerate(asset_vol):
    plt.text(y = v + .005, x = i - .175, s = str(round(v,3)), fontweight='bold');

plt.xticks(rotation = 45)
plt.ylabel('Annualized Standard Deviation');
```

## **Volatility Calculations After Adding Different Asset Classes**

```python
asset_dict = {}

for i in range(len(asset_list)):
  asset_dict.update({', '.join([asset.name for asset in asset_list[:i+1]]) : 252**.5 * (sum([asset for asset in asset_list[:i+1]]).pct_change(1).std())})
```

```python
for title,stdev in asset_dict.items():
  plt.bar(x = title, height = stdev)
  plt.xticks(rotation = 75)

for i, v in enumerate(asset_dict.values()):
    plt.text(y = v + .001, x = i - .2, s = str(round(v,4)), fontweight='bold');

plt.ylabel('Annualized Standard Deviation');
```
