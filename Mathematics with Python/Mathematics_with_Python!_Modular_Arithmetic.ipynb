# **Modular Arithmetic**

**Modular arithmetic is the process where the modulo operation returns the remainder when dividing. The equation below outlines the operation. The numbers within the operation go up to $B$ which is known as the modulus which sets the range of outputs.**

<br/>

$\Large{\frac{A}{B}} $ $= Q \text{ remainder } R$

<br/>

$\text{Where: $A$ is the dividend, $B$ is the divisor (and modulus), $Q$ is the quotient and $R$ is the remainder}$

<sup>Source: [What is modular arithmetic?](https://www.khanacademy.org/computing/computer-science/cryptography/modarithmetic/a/what-is-modular-arithmetic) from Khan Academy<sup/>

# **Modular Arithmetic within Python**

**Within Python, the `%` symbol is used to conduct a modulo operation. The built in modulo expression works on integers and floating numbers. The `fmod` function from Python's math package can also be used to conduct modulo operations. There is a difference between Python `%` function and math's `fmod` function (see the math package documentation for details).**

<sup>Source: [Python Expressions: Modulo Operator](https://docs.python.org/3.3/reference/expressions.html#binary-arithmetic-operations) from python.org<sup/>

## **Modulo Operation**

### **Example 1**

$6 \text{ mod } 2 = 0$ 

$\text{Explanation: } \frac{6}{2} = 6 - (2*3) = 3 \text{ remainder } 0$

<br/>


### **Example 2**

$5 \text{ mod } 2 = 1$ 

$\text{Explanation: } \frac{5}{2} = 5 - (2*2) = 2 \text{ remainder } 1$

```python
print(f'6 mod 2: {6%2}')
print(f'5 mod 2: {5%2}')
```

## **Creating a Modulo Function from Scratch**

**The floor division operator `//` can be utilized to create our modulo function. The operator returns the quotient.**

```python
def modulo(divisor,mod):

  return divisor - (divisor//mod * mod)
```

```python
print(f'6 mod 2: {modulo(6,2)}')
print(f'5 mod 2: {modulo(5,2)}')
```

```python
for div in range(-10,10):
  for modulus in range(1,10):
    print(f'{div} mod {modulus} = {modulo(div,modulus)}', div%modulus == modulo(div,modulus))
```

## **Programming a Clock with Modular Arithmetic**

**We want to program a function in Python to return the hour in military time. Military ranges from 0 (stylized as 0000) up to 23 (2300).**

24hr-clock.jpeg

```python
def clock(hr = 0, tz = 'EST'):

  current_time = datetime.now(pytz.timezone(tz)).hour
  updated_time = (current_time + hr) % 24
  minutes = datetime.now(pytz.timezone(tz)).minute

  if len(str(minutes)) == 1:
    minutes = '0' + str(minutes)
  else:
    minutes = str(minutes)

  if len(str(updated_time)) == 1:
    print(f'Military Time: {"0"+str(updated_time)+minutes}')
  else:
    print(f'Military Time: {str(updated_time)+minutes}')
```

```python
clock(hr = -1)
```

```python
clock(hr=0,tz='Africa/Addis_Ababa')
```

# **Applications of Modular Arithmetic**

**There are many real world applications that use modular arithmetic.**

- **Cryptographic applications such as the Diffie-Hellman key exchange utilize modular arithmetic. This allows two parties to securly exchange cryptographic keys over a public channel.** 

- **Modular Arithmetic is used to track inventory within a system. Universal Product Codes (UPCs) or barcodes on items such as cereal boxes and soup cans help grocery stores track the items in a store. The 10 digits that comprise a UPC are used in a check digit operation that incorporates moduler arithmetic to verify that the UPC is valid (see the article by Professor Ruth Berger for an in-depth explanation).**

- **The International Standard Book Number (ISBN) also uses modular arithmetic to identify books using a unique value (see the article by Professor Ruth Berger for an in-depth explanation).**
