```python
import pandas as pd
```

# **Read the data into a pandas DataFrame**

```python
df = pd.read_csv('nba_player_stats.csv')
df.head()
```

# **Sort DataFrame on a Single Column**

```python
df.sort_values(by='Player')
```

# **Permanently Sort a DataFrame on a Single Column**

```python
#permenately sort a pandas DataFrame by setting "inplace=True"
df.sort_values(by='Player',inplace=True,ignore_index=True)
```

```python
df
```

# **Sort DataFrame on Multiple Columns**

```python
df.sort_values(by=['Team','Player'])
```

# **Sort DataFrame in Descending Order**

```python
df.sort_values(by=['Player'],ascending=False)
```

# **Position Null Values on Sort**

```python
df.sort_values(by=['Team'], na_position='last')
```

# **Sort DataFrame on Key**

```python
df.sort_values(by='Player', key=lambda name: name.str.len())
```
