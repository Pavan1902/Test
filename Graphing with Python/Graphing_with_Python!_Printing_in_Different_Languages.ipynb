```python
import zipfile
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import plotly.express as px

!pip install yfinance
import yfinance as yf

from IPython.core.display import clear_output
clear_output()
```

```python
plt.rcParams["figure.figsize"] = (14,10)
plt.rcParams.update({'font.size': 18})
```

# **Stock Data**

**Download historical prices for the SPY S&P 500 ETF over the past 6 months using `yfinance`**

```python
spy_history = yf.Ticker('SPY').history(period='6mo').dropna()
spy_history
```

# **Hindi Title**

```python
with zipfile.ZipFile('/content/Noto_Sans.zip', 'r') as zip_ref:
  zip_ref.extractall('/content')
```

```python
hindi_font = fm.FontProperties(fname='/content/NotoSans-Black.ttf')

spy_history.plot(y='Close')
plt.title('SPY शेयर की कीमत', fontproperties=hindi_font)
plt.xlabel('दिनांक', fontproperties=hindi_font)
plt.ylabel('कीमत', fontproperties=hindi_font)
plt.show()
```

# **Thai Title**

```python
with zipfile.ZipFile('/content/Kanit.zip', 'r') as zip_ref:
    zip_ref.extractall('/content')

thai_font = fm.FontProperties(fname='/content/Kanit-Black.ttf')
```

```python
spy_history.plot(y='Close')
plt.title('SPY ราคาหุ้น', fontproperties=thai_font)
plt.xlabel('ราคา', fontproperties=thai_font)
plt.ylabel('วันที่', fontproperties=thai_font)
plt.show();
```

# **Plotly**

```python
fig = px.line(spy_history,x=spy_history.index,y='Close', title='SPY शेयर की कीमत',
              labels = dict(Date='दिनांक',Close='कीमत'))

fig.show()
```
